# Cars Dataset

We will Analyze the given dataset which contains the information about cars manufactured by different countries.

```python
import pandas as pd
```

```python
car = pd.read_csv(r"C:\Users\kinsh\Downloads\Cars Data1.csv")
```

```python
car.head()
```

```python
car.shape
```

To check if there are any null values

```python
car.isnull()
```

```python
car.isnull().sum()
```

As we are going to have operations on these attributes therefore we'll replace all the null values with mean

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```

```python
car['EngineSize'].fillna(car['EngineSize'].mean(), inplace = True)
```

```python
car['Length'].fillna(car['Length'].mean(), inplace = True)
```

```python
car['MPG_City'].fillna(car['MPG_City'].mean(), inplace = True)
```

```python
car['MPG_Highway'].fillna(car['MPG_Highway'].mean(), inplace = True)
```

# Average mileage of Cars in both city and highway

Average Mileage in city

```python
car.groupby(['Make']).MPG_City.sum()/car['Make'].value_counts()
```

Average Mileage in Highway

```python
car.groupby(['Make']).MPG_Highway.sum()/car['Make'].value_counts()
```
